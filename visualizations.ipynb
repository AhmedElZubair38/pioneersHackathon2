```python
import pandas as pd

blendstock = pd.read_csv("blendstock.csv")
after = pd.read_csv("after.csv")
```

```python
import matplotlib.pyplot as plt

location_quantities = after.groupby('Destination')['Quantity'].sum()

# Define the explode parameters, with a higher value for the specific locations to offset them
explode_values = [0.1 if location in ['Warehouse', 'Transfer Station A', 'Transfer Station D'] else 0 for location in location_quantities.index]

# Define a custom color map, with distinct colors for the specified locations
highlighted_locations = ['Warehouse', 'Transfer Station A', 'Transfer Station D']
colors = ['#ff9999' if location in highlighted_locations else '#c2c2f0' for location in location_quantities.index]

# Generate the pie chart
plt.figure(figsize=(10, 7))
location_quantities.plot(kind='pie', explode=explode_values, colors=colors, autopct='%1.1f%%')
plt.title('Avg % of Movements to Batch Transfers Before')
plt.ylabel('\n')  # This hides the y-axis label for cleanliness

# Save the figure if needed
plt.savefig('highlighted_pie_chart.png')

# Display the pie chart
plt.show()
```
