```python
import pandas as pd
import numpy as np

post = pd.read_csv('logbookFinal.csv')
blenderStock = pd.read_csv('blendstock.csv')


locations = blenderStock['Location'].values
random_locations = np.random.choice(locations, size=len(post), replace=True)
post['Destination'] = random_locations
post = post.sample(frac=1).reset_index(drop=True)
post['Sequence'] = range(1, len(post) + 1)
post = post.drop(['Task Duration (Minutes)', 'Task Duration (Hours)'], axis=1)
```
