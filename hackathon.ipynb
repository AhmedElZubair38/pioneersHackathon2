```python
import pandas as pd
```

```python
production = pd.read_csv("https://raw.githubusercontent.com/AhmedElZubair38/pioneersHackathon2/main/production.csv")

production
```

```python
rm = pd.read_csv("https://raw.githubusercontent.com/AhmedElZubair38/pioneersHackathon2/main/RM.csv")

rm
```

```python
blendstock = pd.read_csv("https://raw.githubusercontent.com/AhmedElZubair38/pioneersHackathon2/main/blendstock.csv")

blendstock
```

```python
consumption = pd.read_csv("https://raw.githubusercontent.com/AhmedElZubair38/pioneersHackathon2/main/Consumption%20Rates%20%26%20Efficiencies%20of%20Production%20Lines-.csv")

consumption['Ideal Consumption Rate (kg/h)'] = consumption['Ideal Consumption Rate (kg/h)'].str.replace(',', '').astype(int).astype(int)

consumption['Efficiency (%)'] = consumption['Efficiency (%)'].str.replace('%', '').astype(int) / 100

consumption
```

```python
consumption.info()
```

```python
sumi = production[['Blend Code', 'Quantity (kg)']]

sumi
```

```python
blendcode_sums = blendstock.groupby('Blend Code')['Quantity (kg)'].sum().reset_index()

blendcode_sums
```

```python
merged_df = pd.merge(sumi, blendcode_sums, on='Blend Code', how='left', suffixes=('_long', '_short'))

merged_df
```

```python
merged_df['subtracted'] = merged_df['Quantity (kg)_long'] - merged_df['Quantity (kg)_short']

merged_df
```

```python
production['updatedQuantity'] = merged_df['subtracted']

production
```

```python
consumption['Name'] = consumption['Name'].astype(str)

consumption
```

```python
consumption.info()
```

```python
final = production.groupby('Line Name')['updatedQuantity'].sum().reset_index()

final.columns = ['Name', 'Total Quantity (kg)']

final['Name'] = final['Name'].astype(str)

final
```

```python
import pandas as pd

# Assume `df` is your dataframe
# Convert the 'Name' column to string just in case it's not
final['Name'] = final['Name'].astype(str)

# Extract the line number from the 'Name' column
final['Line_Number'] = final['Name'].str.extract('(\d+)', expand=False).astype(int)

# Sort the dataframe by the line number
df_sorted = final.sort_values(by='Line_Number').reset_index(drop=True)

# Drop the 'Line_Number' column if you don't need it anymore
df_sorted = df_sorted.drop(columns=['Line_Number'])

# Display the sorted dataframe
df_sorted
```

```python
consumption["totalAll"] = df_sorted["Total Quantity (kg)"]

consumption
```

```python
consumption['TTC in minutes'] = (consumption['totalAll'] / (consumption['Ideal Consumption Rate (kg/h)'] * consumption['Efficiency (%)'])) * 60

consumption['TTC in hours'] = (consumption['totalAll'] / (consumption['Ideal Consumption Rate (kg/h)'] * consumption['Efficiency (%)']))

consumptionUpdated = consumption
```

```python
# save to csv
consumptionUpdated.to_csv('consumptionUpdated.csv', index=False)
```
