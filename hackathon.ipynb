```python
rm = pd.read_csv("https://raw.githubusercontent.com/AhmedElZubair38/pioneersHackathon2/main/RM.csv")

rm
```

```python
blendstock = pd.read_csv("https://raw.githubusercontent.com/AhmedElZubair38/pioneersHackathon2/main/blendstock.csv")

blendstock
```

```python
consumption = pd.read_csv("https://raw.githubusercontent.com/AhmedElZubair38/pioneersHackathon2/main/Consumption%20Rates%20%26%20Efficiencies%20of%20Production%20Lines-.csv")

consumption['Ideal Consumption Rate (kg/h)'] = consumption['Ideal Consumption Rate (kg/h)'].str.replace(',', '').astype(int).astype(int)

consumption['Efficiency (%)'] = consumption['Efficiency (%)'].str.replace('%', '').astype(int) / 100

consumption
```

```python
consumption.info()
```

```python
unique_blendcode_totals = production.groupby('Blend Code')['Quantity (kg)'].sum().reset_index()

unique_blendcode_totals
```

```python
x = blendstock.groupby('Blend Code')['Quantity (kg)'].sum().reset_index()

updated_blendstock = pd.DataFrame(unique_blendcode_totals['Quantity (kg)'] - x['Quantity (kg)'])

updated_blendstock
```

```python
production_sorted = production.sort_values(by=['Line Name', 'Blend Code', 'Sequence'])

multi_indexed_df = production_sorted.set_index(['Line Name', 'Sequence', 'Blend Code'])

multi_indexed_df
```

```python
total_quantity_by_line = multi_indexed_df.groupby('Line Name')['Quantity (kg)'].sum().reset_index()

total_quantity_by_line
```

```python
hi = blendstock.groupby('Blend Code')['Quantity (kg)'].sum().reset_index()

hi
```

```python
calc = pd.DataFrame(total_quantity_by_line['Quantity (kg)'] - hi['Quantity (kg)'])

calc
```

```python
consumption['TTC'] = (updated_blendstock['Quantity (kg)'] / (consumption['Ideal Consumption Rate (kg/h)'] * consumption['Efficiency (%)']))

consumption
```
